```python
import pandas as pd
import numpy as np
import random

# Step 1: Generate dataset
prior_mean = 5.0
prior_precision = 2.0
likelihood_precision = 4.0

data = []
for i in range(1, 201):
    value = round(random.normalvariate(prior_mean, 1 / (likelihood_precision ** 0.5)), 2)
    data.append([i, value, prior_mean, prior_precision, likelihood_precision])

df = pd.DataFrame(data, columns=["Observation", "Value", "Prior Mean", "Prior Precision", "Likelihood Precision"])

# Save dataset CSV
df.to_csv("bayesian_precision_dataset_200.csv", index=False)
print("Dataset saved as 'bayesian_precision_dataset_200.csv'")

# Step 2: Bayesian updating
posterior_precision = prior_precision + likelihood_precision
df['Posterior Mean'] = (prior_precision * prior_mean + likelihood_precision * df['Value']) / posterior_precision
df['Posterior Precision'] = posterior_precision

# Step 3: Bayesian decision rule (example threshold = 5.0)
threshold = 5.0
df['Decision'] = np.where(df['Posterior Mean'] > threshold, 1, 0)

# Save output CSV
df.to_csv("bayesian_decision_output.csv", index=False)
print("Output saved as 'bayesian_decision_output.csv'")

# Optional: show first 5 rows
print(df.head())
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import random

# Step 1: Generate synthetic 'affairs' dataset
# We'll create 300 samples with 'affairs' feature and binary 'affair_status' label

np.random.seed(42)

n_samples = 300

# Class 0: people with no affairs (majority), low number of affairs near zero
affairs_0 = np.random.poisson(lam=0.2, size=int(n_samples * 0.7))

# Class 1: people who had affairs, higher number of affairs
affairs_1 = np.random.poisson(lam=2.5, size=int(n_samples * 0.3))

affairs = np.concatenate([affairs_0, affairs_1])
labels = np.array([0]*len(affairs_0) + [1]*len(affairs_1))

df = pd.DataFrame({'affairs': affairs, 'affair_status': labels})

print("Synthetic affairs dataset head:")
print(df.head())

# Step 2: Bayesian decision theory

feature_col = 'affairs'
label_col = 'affair_status'

X = df[feature_col].values
y = df[label_col].values

# Calculate priors
prior_0 = np.mean(y == 0)
prior_1 = np.mean(y == 1)

print(f"Prior P(class=0): {prior_0:.3f}")
print(f"Prior P(class=1): {prior_1:.3f}")

# Estimate likelihood parameters for each class (assuming Gaussian for simplicity)
mean_0, std_0 = X[y == 0].mean(), X[y == 0].std(ddof=1)
mean_1, std_1 = X[y == 1].mean(), X[y == 1].std(ddof=1)

print(f"Class 0: mean={mean_0:.3f}, std={std_0:.3f}")
print(f"Class 1: mean={mean_1:.3f}, std={std_1:.3f}")

def posterior_prob(x):
    likelihood_0 = norm.pdf(x, mean_0, std_0)
    likelihood_1 = norm.pdf(x, mean_1, std_1)
    numerator_0 = likelihood_0 * prior_0
    numerator_1 = likelihood_1 * prior_1
    evidence = numerator_0 + numerator_1
    return numerator_0 / evidence, numerator_1 / evidence

posteriors = np.array([posterior_prob(x) for x in X])

predictions = np.argmax(posteriors, axis=1)

accuracy = np.mean(predictions == y)
print(f"Classification accuracy: {accuracy:.4f}")

df['predicted_label'] = predictions

# Save output
df.to_csv("synthetic_affairs_with_predictions.csv", index=False)
print("Results saved to 'synthetic_affairs_with_predictions.csv'")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Values range for plotting
x_values = np.linspace(-1, 8, 500)

# Class conditional likelihoods
likelihood_0 = norm.pdf(x_values, mean_0, std_0)
likelihood_1 = norm.pdf(x_values, mean_1, std_1)

# Priors
prior_0 = np.mean(y == 0)
prior_1 = np.mean(y == 1)

# Posteriors for each x
posterior_0 = (likelihood_0 * prior_0) / (likelihood_0 * prior_0 + likelihood_1 * prior_1)
posterior_1 = (likelihood_1 * prior_1) / (likelihood_0 * prior_0 + likelihood_1 * prior_1)

# Plot class likelihoods
plt.figure(figsize=(10, 6))
plt.plot(x_values, likelihood_0, label='Likelihood class 0 (no affair)', color='blue')
plt.plot(x_values, likelihood_1, label='Likelihood class 1 (affair)', color='red')

# Plot posteriors
plt.plot(x_values, posterior_0, label='Posterior class 0', linestyle='--', color='blue')
plt.plot(x_values, posterior_1, label='Posterior class 1', linestyle='--', color='red')

# Plot decision boundary where posteriors cross (where decision flips)
decision_boundary = x_values[np.argmin(np.abs(posterior_0 - posterior_1))]
plt.axvline(decision_boundary, color='green', linestyle=':', label=f'Decision boundary ≈ {decision_boundary:.2f}')

plt.xlabel('Number of affairs')
plt.ylabel('Probability / Density')
plt.title('Bayesian Decision Theory: Likelihoods and Posterior Probabilities')
plt.legend()
plt.grid(True)
plt.show()
```
