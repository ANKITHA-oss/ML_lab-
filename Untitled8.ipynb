```python
from sklearn.datasets import fetch_openml

# Load MNIST dataset (70,000 samples of 28x28 grayscale digits)
print("Downloading MNIST dataset...")
mnist = fetch_openml('mnist_784', version=1, as_frame=False)

X, y = mnist["data"], mnist["target"].astype(int)
print("MNIST loaded successfully!")
print(f"Data shape: {X.shape}, Labels shape: {y.shape}")
```

```python
# Binary labels: 1 if digit == 0, else 0
y_binary = (y == 0).astype(int)

# Split into training (first 60k) and test (last 10k)
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y_binary[:60000], y_binary[60000:]

print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
from sklearn.linear_model import LogisticRegression

print("Training Logistic Regression model...")
log_reg = LogisticRegression(
    solver='lbfgs',     # good optimizer for this type of problem
    max_iter=1000,      # ensure it converges
)
log_reg.fit(X_train, y_train)
print("Model training complete.")
```

```python
import matplotlib.pyplot as plt

# Show first 10 test images and predictions
y_pred = log_reg.predict(X_test) # Add this line to make predictions

for i in range(10):
    image = X_test[i].reshape(28, 28)
    plt.imshow(image, cmap="gray")
    plt.title(f"Label: {'0' if y_test[i]==1 else 'Not 0'} | Predicted: {'0' if y_pred[i]==1 else 'Not 0'}")
    plt.axis("off")
    plt.show()
```
